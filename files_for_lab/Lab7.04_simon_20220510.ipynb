```python
# 1. One tailed t-test
data = pd.read_csv('machine.txt', sep='\t')
data
```

```python
data.columns = data.columns.str.replace('    ','')
data.columns = data.columns.str.replace(' ','_')
data.columns= data.columns.str.lower()
```

```python
data
```

```python
data['new_machine'].mean()
```

```python
data['old_machine'].mean()
```

```python
# H0: new machine mean time >= old machine mean time 43.23 sec 
# H1: new machine mean time < old machine mean time 43.23 sec 
```

```python
print('p value (single tailed): ',st.ttest_1samp(data['new_machine'],43.23)[1]/2)
```

```python
# 2. Matched Pairs Test

data = pd.read_csv('pokemon.csv')
data
```

```python
data.columns = data.columns.str.replace(' ','_')
data.columns= data.columns.str.lower()
```

```python
pkmn = data[['attack', 'defense']]
pkmn
```

```python
# H0: attack = defense
# H1: attack != defense 
```

```python
st.ttest_rel(pkmn['attack'], pkmn['defense'])
```

```python
# H0 can be recjected
```

```python
data = pd.read_excel('anova_lab_data.xlsx')
data
```

```python
data.columns = data.columns.str.replace(' ','_')
data.columns= data.columns.str.lower()
```

```python
data['power_count'] = data.groupby('power_').cumcount()

data_pivot = data.pivot(index='power_count', columns='power_', values='etching_rate')
data_pivot.columns = ['power_'+str(x) for x in data_pivot.columns.values]
```

```python
data_pivot.columns = data_pivot.columns.str.replace(' ','_')
data_pivot.columns= data_pivot.columns.str.lower()
```

```python
data_pivot
```

```python
# H0 for ANOVA is always that the means of the various groups are the same
# H1 the means of the various groups are not the same
```

```python
st.f_oneway(data_pivot['power_160_w'],data_pivot['power_180_w'],data_pivot['power_200_w'])
```

```python
data_pivot.mean()
```

```python
# H0 can be recjected
```

```python
# DoF = 3-1
# because 3 categories/columns - 1 beacsue this
degrees_freedom = len(data_pivot.columns) - 1
degrees_freedom
```
